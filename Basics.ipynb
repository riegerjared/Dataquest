```python
import csv
import datetime
```

```python
f = open("guns.csv")
data = list(csv.reader(f))
headers = data[0]
data = data[1:]
```

```python
years = [row[1] for row in  data]
year_counts = {}

for year in years:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
        
#year_counts
headers
```

```python
dates = [datetime.datetime(int(row[1]),int(row[2]), 1) for row in data]
dates[0:5]
```

```python
date_counts = {}

for date in dates:
    if date in date_counts:
        date_counts[date] += 1
    else:
        date_counts[date] = 1
        
date_counts
```

```python
sex_counts = {}

for row in data:
    if row[5] in sex_counts:
        sex_counts[row[5]] += 1
    else:
        sex_counts[row[5]] = 1
        
sex_counts
```

```python
race_counts = {}

for row in data:
    if row[7] in race_counts:
        race_counts[row[7]] += 1
    else:
        race_counts[row[7]] = 1
        
race_counts
```

Interesting to see that the number of gun violences recorded is actually higher for white populations. 3 times higher than blacks.
```python
homicide_race_counts = {}

for i,race in enumerate(races) :
    if intents[i] == "Homicide":
        if race in homicide_race_counts:
            homicide_race_counts[race] += 1
        else:
            homicide_race_counts[race] = 1

homicide_race_counts
```

```python
race_per_hundredk_homicide = {}

for race in race_counts:
    race_per_hundredk_homicide[race] = (homicide_race_counts[race]/mapping[race])*100000

race_per_hundredk_homicide   
```
